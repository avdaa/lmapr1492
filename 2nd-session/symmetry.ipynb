```python
from pymatgen.ext.matproj import MPRester
from pymatgen.io.cif import CifParser
from pymatgen.symmetry.analyzer import SpacegroupAnalyzer
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as img
with MPRester("rIpdpPuIE9I2anRpcTZ4") as m:
    # loop_symmetry_equiv_pos_as_xyz
    
    structure = m.get_structure_by_material_id("mp-20194")
    # cif = CifParser("CeO2_mp-20194_symmetrized.cif")
    
    sites = structure.sites
    symmetryOperation = SpacegroupAnalyzer(structure).get_symmetry_operations()
    S1 = symmetryOperation[0]
    # S1.rotation_matrix = [[1., 0., 0.], [0., 1., 0.], [0., 0., 1.]]
    S1.translation_vector[0] = 0.5
    S1.translation_vector[1] = 0.5
    
    S2 = symmetryOperation[11]
    S2.rotation_matrix[0] = [0., -1., 0.]
    # S2.translation_vector = [0., 0., 0.]
    
    S3 = symmetryOperation[1]
    S3.rotation_matrix[0] = [1., 0., 0.]
    S3.rotation_matrix[1] = [0., 1., 0.]
    # S3.translation_vector = [0., 0., 0.]
    
    Ce = [sites[0].a, sites[0].b, sites[0].c]
    O1 = [sites[1].a, sites[1].b, sites[1].c]
    O2 = [sites[2].a, sites[2].b, sites[2].c]
    
    print('Symmétrie 1 :')
    print(S1, '\n')
    print('Position initiale de l atome :', Ce)
    print('Position de l atome après symmétrie :', np.dot(Ce, S1.rotation_matrix.T) + S1.translation_vector)
    plt.figure(figsize = (8,8))
    plt.imshow(img.imread('Symmetry1.png'))
    plt.show()
    
    print('\n')
    print('Symmétrie 2 :')
    print(S2, '\n')
    print('Position initiale de l atome :', O1)
    print('Position de l atome après symmétrie :', np.dot(O1, S2.rotation_matrix.T) + S2.translation_vector)
    plt.figure(figsize = (8,8))
    plt.imshow(img.imread('Symmetry2.png'))
    plt.show()
    
    print('\n')
    print('Symmétrie 3 :')
    print(S3, '\n')
    print('Position initiale de l atome :', O2)
    print('Position de l atome après symmétrie :', np.dot(O2, S3.rotation_matrix.T) + S3.translation_vector)
    plt.figure(figsize = (8,8))
    plt.imshow(img.imread('Symmetry3.png'))
    plt.show()
    
    # i = 0
    # for S in symmetryOperation:
    #     print(i, S.as_xyz_string())
    #     i = i + 1
    
```
